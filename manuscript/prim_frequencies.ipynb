```python
from rdcanon.pubchem_prims import prims as prims2
from rdcanon.drugbank_prims_with_nots import prims as prims3
from rdcanon.np_prims import prims as prims4
import matplotlib.pyplot as plt

sorted_map = sorted(prims2, key=lambda x: prims2[x], reverse=True)


to_plot = [int(prims2[k]) for k in sorted_map if "!" not in k][1:21]
fig,ax = plt.subplots(1,1,figsize=(2.2,2.2), dpi=300)
print(to_plot)
ax.bar(range(len(to_plot)), to_plot, color="grey")
ax.set_xticks(range(len(to_plot)))
ax.set_xticklabels([k for k in sorted_map if "!" not in k][1:21], rotation=90, fontsize=6)
ax.set_yticks([])


plt.savefig("figures/pubchem_freq.png", dpi=300, bbox_inches='tight', pad_inches=0.01)


sorted_map = sorted(prims3, key=lambda x: prims3[x], reverse=True)


to_plot = [int(prims3[k]) for k in sorted_map if "!" not in k][1:21]
fig,ax = plt.subplots(1,1,figsize=(2.2,2.2), dpi=300)
print(to_plot)
ax.bar(range(len(to_plot)), to_plot, color="grey")
ax.set_xticks(range(len(to_plot)))
ax.set_xticklabels([k for k in sorted_map if "!" not in k][1:21], rotation=90, fontsize=6)
ax.set_yticks([])


plt.savefig("figures/drugbank_freq.png", dpi=300, bbox_inches='tight', pad_inches=0.01)


sorted_map = sorted(prims4, key=lambda x: prims4[x], reverse=True)
to_plot = [int(prims4[k]) for k in sorted_map if "!" not in k][1:21]
fig,ax = plt.subplots(1,1,figsize=(2.2,2.2), dpi=300)
print(to_plot)
ax.bar(range(len(to_plot)), to_plot, color="grey")
ax.set_xticks(range(len(to_plot)))
ax.set_xticklabels([k for k in sorted_map if "!" not in k][1:21], rotation=90, fontsize=6)
ax.set_yticks([])


plt.savefig("figures/npatlas_freq.png", dpi=300, bbox_inches='tight', pad_inches=0.01)
```
