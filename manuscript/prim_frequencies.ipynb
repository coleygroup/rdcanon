```python
from rdcanon import canon_reaction_smarts

original = "[c:1][NH:2][C@H:4]([c,CX4!H0:40])[C@:5]([#1:99])([CH2,CH3,O:50])[C:6](=[O:7])[CX4:8]([#1:9])([#1:21])[#6,#1:3].[OH2:10]>>[c:1][N:2].[*:40][C:4]=[O:10].[*:50][C:5]([#1:99])[C:6](=[O:7])[C:8]([#1:9])([#1:21])[*:3]"
unmapped = canon_reaction_smarts(original)
mapped = canon_reaction_smarts(original, True)
remapped = canon_reaction_smarts(original, True, "drugbank", True)

print(original)
print(unmapped)
print(mapped)
print(remapped)
```

```python
from rdkit.Chem import AllChem
from rdkit import Chem

def collect_products(p):
    sets = []
    for i in p:
        for l in i:
            l = Chem.RemoveHs(l)
            sm = Chem.MolToSmiles(l)
            if sm not in sets:
                sets.append(sm)
    return set(sets)


rxn_original = AllChem.ReactionFromSmarts(original)
rxn_unmapped = AllChem.ReactionFromSmarts(unmapped)
rxn_mapped = AllChem.ReactionFromSmarts(mapped)
rxn_remapped = AllChem.ReactionFromSmarts(remapped)

reactant_mol1 = Chem.AddHs(Chem.MolFromSmiles(
    "O=C(C)[C@@H](O)[C@@H](C(C)C)Nc1ccc(OC)cc1"))
reactant_mol2 = Chem.AddHs(Chem.MolFromSmiles("[OH2]"))

print(reactant_mol1.HasSubstructMatch(rxn_original.GetReactants()[0]))
print(reactant_mol2.HasSubstructMatch(rxn_original.GetReactants()[1]))

all_product_sets = []

p = rxn_original.RunReactants((reactant_mol1, reactant_mol2))
all_product_sets.append(collect_products(p))
p = rxn_mapped.RunReactants((reactant_mol1, reactant_mol2))
all_product_sets.append(collect_products(p))
p = rxn_remapped.RunReactants((reactant_mol1, reactant_mol2))
all_product_sets.append(collect_products(p))

print(all_product_sets[0] == all_product_sets[1] == all_product_sets[2])
```
